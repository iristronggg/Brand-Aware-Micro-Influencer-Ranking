# function to get related hashtags

https://www.tagsfinder.com/en-us/ajax/?hashtag=ootd&limit=30&country=us&fs=off&fp=off&fg=off&flike=off&custom=love%20style&type=live

```python
def get_related_hashtag(hashtag, custom_stop, tag_num):
    baseURL = f'https://www.tagsfinder.com/en-us/ajax/?hashtag={hashtag}&limit={tag_num}&country=us&fs=off&fp=off&fg=off&type=live&custom='
    html_response = requests.get(baseURL+urllib.parse.quote(custom_stop))
    soup = BeautifulSoup(html_response.text, 'html.parser')
    tag_elements = soup.find_all('span', class_='tag')
    hashtags = [tag.text[1:-1] for tag in tag_elements][1:]
    all_hashtags = [hashtag] + hashtags
    
#     print(all_hashtags)
    return all_hashtags   
```

```python
def singularize(word):
    # Get the synsets for the word
    synsets = wordnet.synsets(word)

    if synsets:
        # Get the first synset (most common usage)
        synset = synsets[0]
        
        # Get the lemma names (forms) of the synset
        lemma_names = synset.lemma_names()

        if lemma_names:
            # Return the singular form (first lemma name)
            return lemma_names[0]

    # Return the original word if no singular form is found
    return word
```
