### with weight version

```python
with open('concept_extraction/brand_mul_concept_profile_hashtag_id_pair.pkl', 'rb') as f:
    brand_concept_id_pair = pickle.load(f)
with open('concept_extraction/inf_mul_concept_profile_hashtag_id_pair.pkl', 'rb') as f:
    inf_concept_id_pair = pickle.load(f)
```

```python
brand_concept_id_pair_weighted = dict()
inf_concept_id_pair_weighted = dict()
for pair in brand_concept_id_pair:
    brand_concept_id_pair_weighted[pair] = 1
for pair in inf_concept_id_pair:
    inf_concept_id_pair_weighted[pair] = 1
```

```python
for aid in brand_use_hashtags:
    ht_list = brand_use_hashtags[aid]
    for ht in ht_list:
        if ht in ha_to_concept:
            ht_rel_con_set = set(ha_to_concept[ht])
            for con in ht_rel_con_set:
                con_id = train_concept_id_dict[con]
                if (aid, con_id) not in brand_concept_id_pair_weighted:
                    brand_concept_id_pair_weighted[(aid, con_id)] = 0
                brand_concept_id_pair_weighted[(aid, con_id)] += 1
```

```python
rev_train_concept_id_dict = {v: k for k,v in train_concept_id_dict.items()}
```

```python
for pair, weight in brand_concept_id_pair_weighted.items():
    if weight > 100:
        print(all_brands[pair[0]], rev_train_concept_id_dict[pair[1]])
```

```python
for aid in inf_use_hashtags:
    ht_list = inf_use_hashtags[aid]
    for ht in ht_list:
        if ht in ha_to_concept:
            ht_rel_con_set = set(ha_to_concept[ht])
            for con in ht_rel_con_set:
                con_id = train_concept_id_dict[con]
                if (aid, con_id) not in inf_concept_id_pair_weighted:
                    inf_concept_id_pair_weighted[(aid, con_id)] = 0
                inf_concept_id_pair_weighted[(aid, con_id)] += 1
```

```python
with open('concept_extraction/brand_mul_concept_profile_hashtag_id_pair_weighted_0615_r1.pkl', 'wb') as f:
    pickle.dump(brand_concept_id_pair_weighted, f)
with open('concept_extraction/inf_mul_concept_profile_hashtag_id_pair_weighted_0615_r1.pkl', 'wb') as f:
    pickle.dump(inf_concept_id_pair_weighted, f)
    
```

## new (concept-concept) pair using cooccurrence to filter concept relation pair

```python
concept_concept_id_pair = set()
for ht in ha_to_concept:
    ht_rel_con_set = set(ha_to_concept[ht])
    for con1 in ht_rel_con_set:
        con1_id = train_concept_id_dict[con1]
        for con2 in ht_rel_con_set:
            con2_id = train_concept_id_dict[con2]
            if con1 != con2 and (con2_id, con1_id) not in concept_concept_id_pair:
                concept_concept_id_pair.add((con1_id, con2_id))
len(concept_concept_id_pair)
```

```python
concept_pair_score = dict()
for pair in concept_concept_id_pair:
    con1 = pair[0]
    con2 = pair[1]
    score = 0
    for ht in ha_to_concept:
        ht_rel_con_set = set(ha_to_concept[ht])
        if (rev_train_concept_id_dict[con1] in ht_rel_con_set) and (rev_train_concept_id_dict[con2] in ht_rel_con_set):
            score += 1
    concept_pair_score[pair] = score
            
```

```python
len(concept_pair_score)
```

```python
concept_name_pair_score = dict()
for pair in concept_pair_score:
    concept_name_pair_score[(rev_train_concept_id_dict[pair[0]], rev_train_concept_id_dict[pair[1]])] = concept_pair_score[pair]
len(concept_name_pair_score)
```

```python
concept_name_pair_score
```

```python
sorted_concept_name_pair_score = {k: v for k, v in sorted(concept_name_pair_score.items(), key=lambda item: item[1], reverse=True)}
```
