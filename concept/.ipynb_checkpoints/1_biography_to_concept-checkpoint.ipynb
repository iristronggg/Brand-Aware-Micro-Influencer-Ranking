```python
def remove_pattern(input_txt, pattern):
    r = re.findall(pattern, input_txt)
    for i in r:
        input_txt = re.sub(i, '', input_txt)
    return input_txt

def preprocess(df):
    df["text"].fillna("", inplace=True)
    df['tidy_text'] = np.vectorize(remove_pattern)(df['text'], name_regex) 
    df['tidy_text'] = df['tidy_text'].str.replace("[^a-zA-Z]", " ")
    
    print("tokenizing...")
    df['tokenized_text'] = df['tidy_text'].apply(TT.tokenize) # tokenize
    print("stemming...")
    df['tokenized_text'] = df['tokenized_text'].apply(lambda x: [stemmer.stem(i) for i in x]) # stemming
    
    print(df.head(5))
    
    return df
```

```python
def split_count(text):

    emoji_list = []
    data = regex.findall(r'\X', text)
    for word in data:
        if word in emoji.EMOJI_DATA or emoji.demojize(word) != word:
            emoji_list.append(word)
    
    return emoji_list
```

```python
def remove_emoji_from_row(row):
    emoji_list = []
    non_emoji_list = []
    data = regex.findall(r'\X', row)
    for word in data:
        if word in emoji.EMOJI_DATA or emoji.demojize(word) != word:
            emoji_list.append(word)
            non_emoji_list.append(' ')
        else:
            non_emoji_list.append(word)
    return ''.join(non_emoji_list)
```

```python
def remove_stopwords(row):
    return [word for word in row if word.casefold() not in stop_words and len(word)>1]
```

```python
def filter_nouns(row):
    tagged_text = nltk.pos_tag(row)
    nouns = [word for word, tag in tagged_text if tag in tags2]
    return nouns
```

```python
import itertools

def extract_feature(df):
    
    df["biography"].fillna("", inplace=True)
    df["biography"] = df["biography"].str.lower()
    df['mention'] = df['biography'].str.findall(name_regex)
    df['hashtag'] = df['biography'].str.findall(hashtag_regex)
    df['biography'] = df['biography'].str.replace('⠀','')
    df['emoji'] = df['biography'].apply(split_count)
    
    # check the extracted emojis
    print(set(itertools.chain.from_iterable(df['emoji'].values)))
    
    # tokenize text only
    # df['only_text'] = df['biography'].str.replace('\\\\', '\\', regex = True)
    df['only_text'] = df['biography'].replace(name_regex, '', regex = False)
    df['only_text'] = df['only_text'].replace(hashtag_regex, '', regex = True)
    df['only_text'] = df['only_text'].str.replace('\\\\n', ' ', regex = False) # remove line break
    df['only_text'] = df['only_text'].apply(remove_emoji_from_row)
    df['only_text'] = df['only_text'].str.replace("[^a-zA-Z]", " ")  # reomve special characters
    # print(df['only_text'])
    df['tokenized_only_text'] = df['only_text'].apply(nltk.word_tokenize) ## TT.tokenize
    df['tokenized_only_text'] = df['tokenized_only_text'].apply(filter_nouns) ## remain nouns only
    df['tokenized_only_text'] = df['tokenized_only_text'].apply(remove_stopwords)
    
    # count numbers
    # df['#mentions'] = df['mention'].apply(lambda x: len(x))
    # df['#hashtags'] = df['hashtag'].apply(lambda x: len(x))
    # df['#emojis'] = df['emoji'].apply(lambda x: len(x))
    # df['#caption_words'] = df['tokenized_only_text'].apply(lambda x: len(x))
    
    # save final df
    # if brand:
    #     graph_feature = df[['Brand', '#likes', '#comments', '#hashtags', '#mentions', '#emojis', '#caption_words', 'timestamp']]
    # else:
    #     graph_feature = df[['Brand', 'Micro-influencer', '#likes', '#comments', '#hashtags', '#mentions', '#emojis', '#caption_words', 'timestamp']]
    # graph_feature.to_csv(file_name, index=False)
    return df
```

```python
p = inflect.engine()
all_concept_set = set()

def extract_concept_mul(row):
    concept_set = set()
    token_set = set(row)
    if len(token_set) == 0:
        return concept_set
    for tok in token_set:
        print('=====', tok, '=====')
        
        # if it's a plural term, recover it to singular noun
        if p.singular_noun(tok):
            tok = p.singular_noun(tok)
        
        # skip token with high term frequency
        if tok in high_tf_set:
            continue

        if tok in mor_cat_set:  # if token itself is a concept too
            concept_set.add(tok)
            print(tok, "itself is a concept.")
        
        baseURL = 'https://api.conceptnet.io/query?'
        start = f'start=/c/en/{tok}&'
        end = 'end=/c/en&'
        relations = ['rel=/r/IsA', 'rel=/r/RelatedTo', 'rel=/r/FormOf', 'rel=/r/PartOf', 'rel=/r/CapableOf', 'rel=/r/AtLocation', 'rel=/r/CreatedBy',
                     'rel=/r/Synonym', 'rel=/r/DerivedFrom', 'rel=/r/MannerOf', 'rel=/r/SimilarTo', 'rel=/r/EtymologicallyRelatedTo', 'rel=/r/EtymologicallyDerivedFrom']
        limit = '&limit=5'

        for relation in relations:
            try:
                obj = requests.get(baseURL+start+end+relation+limit, proxies=proxy_setting).json()
                
                for edge in obj['edges']:
                    concept = edge['end']['label']
                    # weight = edge['weight']
                    if concept in mor_cat_set:
                        print(relation[7:], concept)
                        concept_set.add(edge['end']['label'])
            except:
                return "timeout"

    print(len(concept_set))
    all_concept_set.update(concept_set)
    
    return concept_set
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
def no_tokenizer(tokens):
    return tokens

def tfidf(texts, keyword_num):

    # vectorizer = TfidfVectorizer(min_df=3, max_df=25, preprocessor=custom_preprocessor, token_pattern=r'[A-Za-z0-9.,?!]+')
    vectorizer = TfidfVectorizer(preprocessor=no_tokenizer, tokenizer=no_tokenizer)
    
    # texts = list(merged_df_brand[merged_df_brand['Brand']==brand]['text'])
#     if len(texts)!=50:
#         print(f'have {len(texts)} posts.')
    
    # Fit and transform the texts using TF-IDF vectorizer
    tfidf_matrix = vectorizer.fit_transform(texts)
    
    # Get feature names (keywords)
    feature_names = vectorizer.get_feature_names_out()
#     print(len(feature_names), feature_names)
    
    # Create a dictionary to store feature names and their corresponding TF-IDF scores
    feature_tfidf_scores = {}
    for i, feature_name in enumerate(feature_names):
        # Get the TF-IDF score for the current feature (word) from the matrix
        tfidf_score = vectorizer.idf_[i]
        feature_tfidf_scores[feature_name] = tfidf_score
    
    # Sort the dictionary keys based on values in descending order
    sorted_kw = sorted(feature_tfidf_scores.keys(), key=lambda x: feature_tfidf_scores[x], reverse=True)[:min(keyword_num, len(feature_names))]
        
#     print(len(sorted_kw), sorted_kw)
    return list(sorted_kw)
```

```python
with open('mor_cat_set.pkl', 'rb') as f:
    mor_cat_set = pickle.load(f)
with open('high_tf_set.pkl', 'rb') as f:
    high_tf_set = pickle.load(f)
    
p = inflect.engine()
all_concept_set = set()

def extract_concept_mul(row):
    concept_set = set()
    token_set = set(row)
    if len(token_set) == 0:
        return concept_set
    for tok in token_set:
        print('=====', tok, '=====')
        
        # if it's a plural term, recover it to singular noun
        if p.singular_noun(tok):
            tok = p.singular_noun(tok)
        
        # skip token with high term frequency
        if tok in high_tf_set:
            continue

        if tok in mor_cat_set:  # if token itself is a concept too
            concept_set.add(tok)
            print(tok, "itself is a concept.")
        
        baseURL = 'https://api.conceptnet.io/query?'
        start = f'start=/c/en/{tok}&'
        end = 'end=/c/en&'
        relations = ['rel=/r/IsA', 'rel=/r/RelatedTo', 'rel=/r/FormOf', 'rel=/r/PartOf', 'rel=/r/CapableOf', 'rel=/r/AtLocation', 'rel=/r/CreatedBy',
                     'rel=/r/Synonym', 'rel=/r/DerivedFrom', 'rel=/r/MannerOf', 'rel=/r/SimilarTo', 'rel=/r/EtymologicallyRelatedTo', 'rel=/r/EtymologicallyDerivedFrom']
        limit = '&limit=5'

        for relation in relations:
            try:
                obj = requests.get(baseURL+start+end+relation+limit, proxies=proxy_setting).json()
                
                for edge in obj['edges']:
                    concept = edge['end']['label']
                    # weight = edge['weight']
                    if concept in mor_cat_set:
                        print(relation[7:], concept)
                        concept_set.add(edge['end']['label'])
            except:
                return "timeout"

    print(len(concept_set))
    all_concept_set.update(concept_set)
    
    return concept_set
```
